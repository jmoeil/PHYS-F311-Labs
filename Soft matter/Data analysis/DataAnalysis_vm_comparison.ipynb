```python
import numpy as np
import matplotlib.pyplot as plt

vm_dmpc = np.array([1140.703938704968,1193.806168122055])
vm_dppc = np.array([1190.8078098035255, 1268.6779581574476])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(13, 6))

ax.plot(['Gel phase','Liquid phase'],vm_dmpc,'o',label='Volume per molecule for DMPC')
ax.plot(['Gel phase','Liquid phase'],vm_dppc,'o',label='Volume per molecule for DPPC')
ax.set_xlabel('Phase considered')
ax.set_ylabel('$v_m$  $[Ã…^3]$')
ax.set_title('Comparison of mean volume per molecule for different phospholipids at different phases')
ax.legend()
```
